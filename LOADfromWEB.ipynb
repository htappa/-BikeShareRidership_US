```python
import pandas as pd
import numpy as np
from io import BytesIO
from zipfile import ZipFile
from urllib.request import urlopen
import glob
```

```python
filelist = glob.glob('*.zip')
dflist = []
colnames = ['duration_min', 'start_time', 'end_time', 'start_station_id', 'start_station_name', 'end_station_id', 
            'end_station_name', 'bike_id', 'user_type']
for filename in filelist :
    url = urlopen('https://s3.amazonaws.com/hubway-data/index.html')
    zipfile = ZipFile(BytesIO(url.read()))
    df = pd.read_csv(zipfile.open(filename, header = None, skiprows = 1))
    print(filename)
    dflist.append(df)
print('READ COMPLETE')

cbs = pd.concat(dflist, axis = 0)
cbs.columns = colnames
print('CONCAT COMPLETE')

#cbs['duration_min'] = cbs['duration_min'].apply(lambda x: x/60) # convert duration (seconds) into duration (minutes)
#cbs = cbs[['bike_id', 'start_time', 'end_time', 'duration_min', 'start_station_id', 'end_station_id', 
#           'start_station_name', 'end_station_name', 'user_type']]
#cbs['count'] = 1
#print('FORMAT COMPLETE')
```
