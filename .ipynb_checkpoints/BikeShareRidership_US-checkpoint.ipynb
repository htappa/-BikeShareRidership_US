# Bike Share Ridership in the United States

The purpose of this code is to analyze and compare the trip data from three different bike share services: Capital Bikeshare, Hubway, and Divvy. The steps are outlined below.

Load packages. This analysis relies heavily on Plotly, and it is recommended that you use your own username and api key for the Plotly sign in.

```python
import pandas as pd
import os
import glob
import plotly.plotly as py
from plotly.graph_objs import *
py.sign_in('htappa', '4Nz4NT3BA74zZkB2ECEg')

pd.options.display.float_format = '{:.0f}'.format
```

Build pandas dataframe for Capital Bikeshare trips in Washington, DC. Since there are 12 files, 4 for each year from 2015 - 2017, the code includes a for-loop that reads in each file and concatenates them into a single dataframe.

```python
os.chdir('capitalbikeshare')

filelist = glob.glob('*.csv')
dflist = []
colnames = ['duration_min', 'start_time', 'end_time', 'start_station_id', 'start_station_name', 'end_station_id', 
            'end_station_name', 'bike_id', 'user_type']
for filename in filelist :
    print(filename)
    df = pd.read_csv(filename, header = None, skiprows = 1)
    dflist.append(df)
print('READ COMPLETE')

cbs = pd.concat(dflist, axis = 0)
cbs.columns = colnames
print('CONCAT COMPLETE')

cbs['duration_min'] = cbs['duration_min'].apply(lambda x: x/60) # convert duration (seconds) into duration (minutes)
cbs = cbs[['bike_id', 'start_time', 'end_time', 'duration_min', 'start_station_id', 'end_station_id', 
           'start_station_name', 'end_station_name', 'user_type']]
cbs['count'] = 1
print('FORMAT COMPLETE')

os.chdir('..')
```

Build pandas dataframe for Hubway trips in Boston, MA. There are 35 files comprising all trips made from 2015 - 2017. Once again, a for-loop reads in each file and concatenates them into a single dataframe. NOTE that there is no publicly-available Hubway trip history data for December 2016, so this data is missing from the Hubway dataframe.

```python
os.chdir('hubway')

filelist = glob.glob('*.csv')
dflist = []
colnames = ['duration_min', 'start_time', 'end_time', 'start_station_id', 'start_station_name', 'start_station_latitude', 
            'start_station_longitude', 'end_station_id', 'end_station_name', 'end_station_latitude', 'end station longitude', 
            'bike_id', 'user_type', 'birth_year', 'gender']
for filename in filelist :
    print(filename)
    df = pd.read_csv(filename, header = None, skiprows = 1)
    dflist.append(df)
print('READ COMPLETE')

hubway = pd.concat(dflist, axis = 0)
hubway.columns = colnames
print('CONCAT COMPLETE')

hubway['duration_min'] = hubway['duration_min'].apply(lambda x: x/60) # convert duration (seconds) into duration (minutes)
hubway = hubway.drop(hubway.columns[[5, 6, 9, 10]], axis = 1)
hubway = hubway[['bike_id', 'start_time', 'end_time', 'duration_min', 'start_station_id', 'end_station_id', 
               'start_station_name', 'end_station_name', 'user_type', 'gender', 'birth_year']]
hubway['gender'] = hubway['gender'].replace(1, 'Male').replace(2, 'Female')
hubway['user_type'] = hubway['user_type'].replace('Subscriber', 'Member').replace('Customer', 'Casual')
hubway['count'] = 1
print('FORMAT COMPLETE')

os.chdir('..')
```

Below is the code used to create the pandas dataframe for Divvy trips in Chicago, IL. Since the start_time variable is used later on to group the obervations by month, and the divvy csv files are not in the standard datetime format required for this aggregation method, a lambda operator was incorporated to convert start_time values to the correct format. Note that this can take over an hour to run.

```python
os.chdir('divvy')

filelist = glob.glob('*.csv')
dflist = []
colnames = ['trip_id', 'start_time', 'end_time', 'bike_id', 'duration_min', 'start_station_id', 'start_station_name', 
            'end_station_id', 'end_station_name', 'user_type', 'gender', 'birth_year']
for filename in filelist :
    print(filename)
    df = pd.read_csv(filename, header = None, skiprows = 1, parse_dates = [1], 
                     date_parser = lambda x: pd.to_datetime(x))
    dflist.append(df)
print('READ COMPLETE')

divvy = pd.concat(dflist, axis = 0)
divvy.columns = colnames
print('CONCAT COMPLETE')

divvy['duration_min'] = divvy['duration_min'].apply(lambda x: x/60)
divvy = divvy.drop(divvy.columns[[0]], axis = 1)
divvy = divvy[['bike_id', 'start_time', 'end_time', 'duration_min', 'start_station_id', 'end_station_id', 
               'start_station_name', 'end_station_name', 'user_type', 'gender', 'birth_year']]
divvy['user_type'] = divvy['user_type'].replace('Subscriber', 
                                                'Member').replace('Customer', 'Casual').replace('Dependent', 'Casual')
divvy['count'] = 1
print('FORMAT COMPLETE')

os.chdir('..')
```
